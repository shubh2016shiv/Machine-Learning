# Data Wrangling

Cleaning the data by removing the NaN values and unnecessary columns

```python
titanic_data.isnull().head(20)
```

```python
titanic_data.isnull().sum()
```

```python
sns.heatmap(titanic_data.isnull(),robust=True)
```

```python
titanic_data.drop('Cabin',inplace=True,axis=1)
```

```python
titanic_data.dropna(inplace=True)
```

```python
titanic_data.isnull().sum()
```

```python
titanic_data.head(10)
```

'Pclass','Name','Sex','Ticket','Embarked' are either Categorical data or string values which are unnecessary, hence remove them. Categorical data will be first converted to dummy values, concatenate and then, parent columns would be removed

```python
sex = pd.get_dummies(titanic_data['Sex'],drop_first=True)
```

```python
embarked = pd.get_dummies(titanic_data['Embarked'],drop_first=True)
pclass = pd.get_dummies(titanic_data['Pclass'],drop_first=True)
```

```python
titanic_data=pd.concat([titanic_data,sex,embarked,pclass],axis=1)
```

```python
titanic_data
```

```python
titanic_data.drop(['PassengerId','Pclass','Name','Sex','Ticket','Embarked'],axis=1,inplace=True)
```

```python
clean_data = titanic_data
```

```python
clean_data.head()
```

# Training And Prediction using Logistic Regression

```python
# Extracting x and y values
y = clean_data['Survived']
x = clean_data.drop(['Survived'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.3, random_state = 20)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(xTrain,yTrain)
```

```python
yPred = model.predict(xTest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(yTest,yPred)
```

```python
accuracy_score(yTest,yPred)
```

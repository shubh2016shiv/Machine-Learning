```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 9)
import matplotlib.style as style
style.use("ggplot")
```

```python
import pandas as pd
```

```python
data = pd.read_csv('cluster_data.csv')
```

```python
data.head()
```

```python
x1 = data['x']
x2 = data['y']
x = np.array([x1,x2]).T
```

```python
plt.scatter(x1,x2,s=50,c='k')
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(x)
```

```python
centroids = kmeans.cluster_centers_
```

```python
labels = kmeans.labels_
```

```python
colors = ['r', 'g', 'b']
fig, ax = plt.subplots()
for i in range(3):
    points = np.array([x[j] for j in range(len(x)) if labels[j] == i])
    ax.scatter(points[:, 0], points[:, 1], s=15, c=colors[i])
plt.scatter(centroids[:,0],centroids[:,1],s=100,marker='x')
```

```python
x_test_1 = float(input("Enter X Value: "))
x_test_2 = float(input("Enter Y Value: "))
```

```python
predict_input = np.array([[x_test_1,x_test_2]])
predict_input
```

```python
predicted_point = kmeans.predict(predict_input)
predicted_point[0]
```

```python
colors = ['r', 'g', 'b']
fig, ax = plt.subplots()
for i in range(3):
    points = np.array([x[j] for j in range(len(x)) if labels[j] == i])
    ax.scatter(points[:, 0], points[:, 1], s=15, c=colors[i])
plt.scatter(centroids[:,0],centroids[:,1],s=100,marker='x')
plt.scatter(predict_input[:,0],predict_input[:,1],s=200,marker='*',c=colors[predicted_point[0]])
```

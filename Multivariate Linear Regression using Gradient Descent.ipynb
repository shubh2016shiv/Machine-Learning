```python
data = pd.read_csv('TwoVariableData.txt',names=["size","bedroom","price"])
```

```python
size = data['size'].values
bedroom = data['bedroom'].values
price = data['price'].values
```

```python
def compute_cost(x_values,y_values,weights):
    cost = (np.sum((np.dot(x_values,weights)-y_values)**2))/len(x_values)
    return cost
```
